```python
#Calculate the percentange of passing math score for each school
grouped_passed_math = school_data_combined[school_data_combined['math_score']>=70].groupby(['school_name']).size()
percent_passed_math = (grouped_passed_math/school_size)*100
```

```python
school_summary= pd.DataFrame(school_summary_data)
```

```python
# Copied data frame
display_school_summary = school_summary.copy()
# Format 
display_school_summary['Per Student Budget'] = display_school_summary['Per Student Budget'].map('${:,.2f}'.format)
display_school_summary['Total School Budget'] = display_school_summary['Total School Budget'].map('${:,.2f}'.format)
display_school_summary.index.name = None
```

```python
# Average math score for 9th grade students at each school
avg_math_9 = school_data_combined[school_data_combined['grade']=='9th'].groupby('school_name')['math_score'].mean()
# Average math score for 10th grade students at each school
avg_math_10 = school_data_combined[school_data_combined['grade']=='10th'].groupby('school_name')['math_score'].mean()
# Average math score for 11th grade students at each school
avg_math_11 = school_data_combined[school_data_combined['grade']=='11th'].groupby('school_name')['math_score'].mean()
# Average math score for 12th grade students at each school
avg_math_12= school_data_combined[school_data_combined['grade']=='12th'].groupby('school_name')['math_score'].mean()

# Results Dataframe by school
math_score_gradelevel={
    '9th':avg_math_9,
    '10th':avg_math_10,
    '11th':avg_math_11,
    '12th':avg_math_12,
    }

math_score_by_gradelevel = pd.DataFrame(math_score_gradelevel)
math_score_by_gradelevel.index.name = None
math_score_by_gradelevel.head(20)
```

```python
# Average reading score for 9th grade students at each school
avg_read_9 = school_data_combined[school_data_combined['grade']=='9th'].groupby('school_name')['reading_score'].mean()
# Average reading score for 10th grade students at each school
avg_read_10 = school_data_combined[school_data_combined['grade']=='10th'].groupby('school_name')['reading_score'].mean()
# Average reading score for 11th grade students at each school
avg_read_11 = school_data_combined[school_data_combined['grade']=='11th'].groupby('school_name')['reading_score'].mean()
# Average reading score for 12th grade students at each school
avg_read_12= school_data_combined[school_data_combined['grade']=='12th'].groupby('school_name')['reading_score'].mean()

# Results Dataframe by school
read_score_gradelevel={
    '9th':avg_read_9,
    '10th':avg_read_10,
    '11th':avg_read_11,
    '12th':avg_read_12,
    }

read_score_by_gradelevel = pd.DataFrame(read_score_gradelevel)
read_score_by_gradelevel.index.name = None
read_score_by_gradelevel.head(20)
```

```python
spending_bin_student = [0, 585, 615, 645, 675]
spending_groups = ["<$585", "$585-615", "$615-645", "$645-675"]


scores_by_spending = school_summary.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','Overall Passing Rate',]]

# Spending Ranges (Per Student) and bin by budget per studnet
scores_by_spending['Spending Ranges (Per Student)']= pd.cut(school_summary['Per Student Budget'],spending_bin_student,labels=spending_groups)


scores_by_spending = scores_by_spending.groupby('Spending Ranges (Per Student)').mean()
scores_by_spending.head()
```

```python
school_size_bins = [0, 1000, 2000, 5000]
school_size_groups = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

scores_by_size = school_summary.loc[:,['Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','Overall Passing Rate',]]

# School Size and bin by total students
scores_by_size['School Size']= pd.cut(school_summary['Total Students'],school_size_bins,labels=school_size_groups)

scores_by_size = scores_by_size.groupby('School Size').mean()
scores_by_size.head()
```

```python
scores_by_type = school_summary[['School Type','Average Math Score',
                                  'Average Reading Score','% Passing Math',
                                  '% Passing Reading','Overall Passing Rate',]]
# Scores by School Type
scores_by_type = scores_by_type.groupby('School Type').mean()
scores_by_type.head()
```

```python
Analysis
-Charter schools have a statistically significant higher passing rate than District school. Although other factors could be involved (such as school size), as a whole charter schools outperform district schools.
- Smaller and medium schools have a statistically higher passing rate  (94% and 95% versus 76%) than Larger (2,000 to 5,000) schools. 
- More budget per student doesn't necessarily correlate to higher passing rates as schools with the most budget per student ($645 to $675) had the lowest overall passing rate (73%)
```
